```python
import numpy as np
import pandas as pd
import os

# Specify git executable file for GitPython in Jupyter Notebook (In IDE, it can still work without this line.)
os.environ["GIT_PYTHON_GIT_EXECUTABLE"] = "C:\Program Files\Git\cmd\git.exe"

import git
from git import RemoteProgress

from git import Repo
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

class Progress(RemoteProgress):
    def update(self, op_code, cur_count, max_count=None, message=''):
        print(self._cur_line)
        
remote_link = "https://github.com/apache/groovy"
local_link = "repo/apache"
# Uncomment to clone
Repo.clone_from(remote_link, local_link, progress=Progress())

repo = Repo(local_link)
fixing_commit = "4df8b652aa018a5d5d1cda8fba938bf3422db31c"
affected_file = "src/main/org/codehaus/groovy/runtime/MethodClosure.java"
```

```python
#How many total files were affected in the fixing commit? 
#Hint: git diff or git show
diff_data = repo.git.diff(fixing_commit + "^", fixing_commit).splitlines()
for line in diff_data:
    print(line)
```

```python
blame_data = repo.git.blame("-L33,+1", fixing_commit + "^", "--", affected_file).splitlines()
for line in blame_data:
    print(line)
vcc_commit = blame_data[0].strip().split()[0]
log_data = repo.git.log("--oneline", vcc_commit + ".." + fixing_commit + "^").splitlines()
print("Number of commits in between:", len(log_data))
```

```python
#a.	What was the message and title of the fixing commit? 
#Was there any mention of fixing a bug or vulnerability? 
#Hint: git diff or git show
show_data = repo.git.show("-s",vcc_commit).splitlines()
for line in show_data:
    print(line)
```

```python
#b.	How many total files were affected in the vcc commit? 
#Hint: git diff or git show
diff_data = repo.git.diff(vcc_commit + "^", vcc_commit)
type(diff_data)
number_files = diff_data.count("diff --git")
print(number_files)
```

```python
#c.	How many total directories were affected in the fixing commit? 
#For example, if a file path is: abc/def/File.java, 
#then its directory is abc/def. Hint: git diff or git show
diff_data = repo.git.diff(vcc_commit + "^", vcc_commit).splitlines()
for line in diff_data:
    print(line)
```

```python
#d.	How many total lines of code (including comments and blank lines) \
#were deleted? Hint: git diff or git show
diff_data = repo.git.diff(vcc_commit + "^", vcc_commit).splitlines()
del_lines = - number_files
for line in diff_data:
    if line[0] == '-':
        del_lines = del_lines + 1
print("Total files deleted")
print(del_lines)
```

```python
#e.	How many total lines of code (including comments and blank lines) 
#were added? Hint: git diff or git show
diff_data = repo.git.diff(vcc_commit + "^", vcc_commit).splitlines()
added_lines = - number_files
for line in diff_data:
    if line[0] == '+':
        added_lines = added_lines + 1
print("Total files added")
print(added_lines)
```

```python
#f.	How many total lines of code (excluding comments and blank lines) 
#were deleted? Hint: git diff or git show
diff_data = repo.git.diff(vcc_commit + "^", vcc_commit).splitlines()
del_lines_excomment = 0
for line in diff_data:
    if line[0] == '-':
        if "//" in line:
            del_lines_excomment = del_lines_excomment + 1
print("Total files deleted excluding comments")
print(del_lines - del_lines_excomment)
```

```python
#g.	How many total lines of code (excluding comments and blank lines) 
#were added? Hint: git diff or git show
diff_data = repo.git.diff(vcc_commit + "^", vcc_commit).splitlines()
added_lines_excomment = 0
for line in diff_data:
    if line[0] == '+':
        if "//" in line:
            added_lines_excomment = added_lines_excomment + 1
print("Total files added excluding comments")
print(added_lines - added_lines_excomment)
```

```python
affected_file = "src/main/org/codehaus/groovy/runtime/MethodClosure.java"
#h.	How many days were between the current fixing commit and the previous 
#commit of each affected file? Hint: git log
#Commit
commit_search = "commit " + "51ae3d70057" 
counter1 = 21
#AffectedFile
log_data = repo.git.log("--parents","--",affected_file).splitlines()
for line in log_data:
    if commit_search in line:
        counter1 = 0
    if counter1 < 15:
        print(line)
        counter1 = counter1 + 1
#i.	How many time has each affected file of the current fixing commit 
#been modified in the past since their creation (including rename of the file)? Hint: git log
#Affected File1
log_data = repo.git.log("--oneline","--",affected_file).splitlines()
print("Number of commits in between:", len(log_data))
#j.	Which developers have modified each affected file since its creation? 
#Hint: git log
log_data = repo.git.log("--pretty=format:%an","--",affected_file).splitlines()
df = pd.DataFrame(log_data, columns=['Author'])
df['count'] = 1
answer1 = df.groupby(['Author']).count().reset_index(0, drop=False)
print(answer1)
```

```python
#k.	For each developer identified, how many commits have each of them 
#submitted? From your observation, are the involving developers
#experienced (with many commits) or new ones (with few commits) or both? 
#Hint: git log or git shortlog
commit_authors = repo.git.shortlog("-sne", "--all").splitlines()

author_commits = []

for commit_author in commit_authors:
    commit, author = commit_author.split("\t")
    author_commits.append([int(commit), author])
df = pd.DataFrame(author_commits, columns=['Commit', 'Author'])
print(df)
```
